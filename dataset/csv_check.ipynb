```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv("data1.csv")
data1['date'] = data1['date'].apply(lambda x: x.split(',')[1])
data1.sort_values("userscore", axis = 0, ascending = False, inplace = True)
data1 = data1[data1.userscore != "tbd"]
data1.set_index(["userscore"], inplace = True)
data1
```

```python
# read data2
data2 = pd.read_csv("data2.csv")

# changing nan in data into 0 and float to int
data2['Year_of_Release'] = data2['Year_of_Release'].fillna(0).apply(np.int64)
data2['Critic_Count'] = data2['Critic_Count'].fillna(0).apply(np.int64)
data2['User_Count'] = data2['User_Count'].fillna(0).apply(np.int64)

# drop publisher, genre, rating, developer
data2 = data2.drop('Publisher',axis = 1)
data2 = data2.drop('Genre',axis = 1)
data2 = data2.drop('Rating',axis = 1)
data2 = data2.drop('Developer',axis = 1)

# drop user < 50 and critic < 10 count
# drop global sales < 0.1 m
user_credible = data2['User_Count'] >50.0
critic_credible = data2['Critic_Count']>10.0
sales_credible = data2['Global_Sales']>0.1
data2_credible = data2[user_credible & critic_credible&sales_credible]

data2_credible
```
