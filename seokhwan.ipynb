### Data3

```python
data3 = pd.read_csv("/Users/s3park2022/Desktop/data_science_project/dataset/data3_modified.csv")

# drops the column 'Unnamed: 0'
data3 = data3.drop("Unnamed: 0", axis = 1)

# removes 'm' in keydata values
data3.keydata = [data3.keydata[i][:-1] for i in range(len(data3))]

# removes unnecessary values in keydata
data3 = data3[data3['keydata'] != "0.00"]

# removes N/A values in release_Date column
data3.drop(data3[data3['release_Date'].str.strip() == "N/A" ].index , inplace = True)

# only leaves the year in release_Date values
for i in range(len(data3)):
    tmp = data3.iloc[i].release_Date.split()[2]
    if (int(tmp) > 19):
        data3.release_Date[i] = "19" + tmp
    elif (int(tmp) < 20):
        data3.release_Date[i] = "20" + tmp

# renames the column name 'title' to 'name'
data1 = data1.rename(columns = {'title':'name'})
        
data3
```
